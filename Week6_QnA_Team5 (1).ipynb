# Q2) Examine the data by using visualizations.

```python
plt.figure(figsize=(16,10))
plt.title('Distribution of Agricultural production index  w.r.to region')
sns.boxplot(x='Region',y='Agricultural production index ',data=data)
plt.xticks(rotation=90)
plt.show()
```

Insights:<br>1. Oceania has the most consistent performance in terms of Agricultural production index;while Western Asia has the widest range.<br>2. South America has the highest median while Southern Europe has the lowest median in terms of Agricultural production index.

```python
plt.figure(figsize=(16,8))
plt.title('Region v/s Population in thousands (2017)')
fig=sns.barplot(x='Region',y='Population in thousands (2017)',ci=None,data=data)
plt.xticks(rotation=90)
plt.ylim(0,350000)
#to get the count displayed on the chart
for p in fig.patches:
    fig.annotate(round(p.get_height(),2) , (p.get_x()+0.3, p.get_height()+10000),rotation=90)
plt.show()
```

Insights:<br>1. Eastern Asia is the most highly populated region while Micronesia is the least populated.

```python
plt.figure(figsize=(16,8))
plt.title('Region v/s Surface area (km2)')
fig=sns.barplot(x='Region',y='Surface area (km2)',ci=None,data=data)
plt.xticks(rotation=90)
plt.ylim(0,6500000)
#to get the count displayed on the chart
for p in fig.patches:
    fig.annotate(round(p.get_height(),2) , (p.get_x()+0.3, p.get_height()+65000),rotation=90)
plt.show()
```

Insights:<br>1. Northern America has the highest surface area while Micronesia has the least area.

# Q4 - a) On average which region has the highest GDP per capita (current US$)

```python
d1 = data.groupby('Region')['GDP per capita (current US$)'].mean()
d1.sort_values(ascending = False).head(1)
```

```python
plt.figure(figsize=(16,6))
plt.title('Region v/s GDP per capita (current US$)')
fig=sns.barplot(x='Region',y='GDP per capita (current US$)',ci=None,data=data)
plt.xticks(rotation=90)
plt.ylim(0,75000)
#to get the count displayed on the chart
for p in fig.patches:
    fig.annotate(round(p.get_height(),2) , (p.get_x()+0.3, p.get_height()+1500),rotation=90)
plt.show()
```

# Q4 - b) On average which region has the highest International trade: Imports (million US$)


```python
d2 = data.groupby('Region')['International trade: Imports (million US$)'].mean()
d2.sort_values(ascending = False).head(1)
```

```python
plt.figure(figsize=(16,8))
plt.title('Region v/s International trade: Imports (million US$)')
fig=sns.barplot(x='Region',y='International trade: Imports (million US$)',ci=None,data=data)
plt.xticks(rotation=90)
plt.ylim(0,780000)
#to get the count displayed on the chart
for p in fig.patches:
    fig.annotate(round(p.get_height(),2) , (p.get_x()+0.3, p.get_height()+10000),rotation=90)
plt.show()
```
